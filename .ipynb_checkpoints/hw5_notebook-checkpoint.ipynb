# Dataset used for this Home Work
Since, the microblog messages are related to the spread of an epidemic I found a dataset of tweets related to the H1N1 (swine flu) outbreak of 2009 (http://cmci.colorado.edu/~mpaul/downloads/flu_data.php). The tweets were of two types:
* tweets by those affected personally by flu-realed symptoms (type-1)
* tweets by those not personally affected by flu-related symptoms (type-2)


In the VAST challenge there are two primary clusters with each cluster having tweets about a different flu. Solving the challenge would involve finding the clusters in the dataset and analyzing the tweets arising from each cluster.

The tweets I obtained did not have GPS and time stamp information so I found a separate GPS dataset i.e. the Mopsi dataset which included GPS coordinates from the region of Joensuu (https://cs.joensuu.fi/sipu/datasets/) that were clustered together in 2 clusters. Since there are 2 clusters and 2 categories of tweets in the dataset I found, I then performed clustering to obtain the clusters in the dataset. I then associated type-1 tweets with the first cluster and type-2 tweets with the second cluster. 


# Machine Learning Algorithms (ML) Used
# ML Algorithm 1: **Clustering**
Solving the challenge would involve finding the clusters of the GPS data in the microblog dataset and analyzing the microblog messages in each cluster. I have performed clustering on the Mopsi dataset using k-means clustering. The code and the results of this effort can be found below. The clusters formed have been saved as numpy arrays  in the folder cluster_data as cluster1.npy and cluster2.npy.

**Analysis:**
On anayzing the results I found that there were very few points in the second cluster. This could be due to the fact that there were many outliers (as seen in the image from the webpage  of the map and all the points in the dataset I obtained) and this may have distoted the results

```python
#Code to extract GPS data from file and perform clustering on it.
import numpy as np
from sklearn.cluster import KMeans, DBSCAN

GPS_data=[]

#Read in the GPS data from the file and store each coordinate as a record in a list
with open('GPS_data/GPS_data.txt','r') as f:
    for line in f:
        record=[float(line.split()[0]), float(line.split()[1])]
        GPS_data.append(record)

#Convert GPS data into a numpy array and perform clustering on it
X = np.array(GPS_data)
kmeans = KMeans(n_clusters=2, random_state=0).fit(X)
cluster_labels=kmeans.labels_


#merge cluster labels with their respective GPS coordinates
cluster_labels=cluster_labels.reshape(-1,1)
GPS_cluster_info=np.hstack([X,cluster_labels])


cluster1_points=[]
#store  points belonging to cluster 1 cluster1_points=[]
for index in range(len(GPS_cluster_info)):
    if GPS_cluster_info[index][2] == 0:
        cluster1_points.append(GPS_cluster_info[index])

np.save('cluster_data/cluster1',np.array(cluster1_points))
print("The GPS-cordinates in the first cluster are")
print(np.array(cluster1_points)[:,0:2])

cluster2_points = []
# store  points belonging to cluster 1 cluster1_points=[]
for index in range(len(GPS_cluster_info)):
    if GPS_cluster_info[index][2] == 1:
        cluster2_points.append(GPS_cluster_info[index])

np.save('cluster_data/cluster2',np.array(cluster2_points))
print("The GPS-cordinates in the second cluster are ")
print(np.array(cluster2_points)[:,0:2])
```

# ML Algorithm 2 : Word Frequency Analysis of microblog messages/tweets in each category

After performing clustering of the points based on the GPS data the microblog messages need to be analyzed to see the kind of tweets arising from each region.  
The assumption made here is that in each category words  that occur more frequencty are more important for that particular category.  
I used sklearns CountVectorization class to find the top 20 most freqeunt words that occur in each category of tweets. Stop words in English such as 'the', 'and', 'for', 'is', 'or' etc. have been excluded because they occur frequently regarless of the category and are not good distinguishing between the two categories of tweets (i.e.  tweets by those affected personally by flu-realed symptoms,
tweets by those not personally affected by flu-related symptoms).  
The Codes and results for text analysis of each Categorty of tweets is as follows:

**Analysis of Results:**
Many of the top 20 words returned (flu, shot, ) from each category are quite similar because the tweets in both cases are related to flu but in one case the tweets relate to personal illness of the person tweeting and in the second case, the tweets are people who are just commenting on the current spread of the flu-epidemic. 
An interesting observation is that tweets from people affected with illness (category-1) have more positive words since we can spot words such as hope, better, like and a few others which are not really found in the other category.



**Conclusion**: 
Because of insufficent time I could not explore other techniques such TF-IDF frequency which could help in reducing the similarity of top 20 tweets obtained in the 2 categories but elimiating words that occur frequently across all tweets in both catgories such as 'flu' which could ultimately give us better analysis.

```python
from sklearn.feature_extraction.text import CountVectorizer

#read the flu teets (directly related to personal sickness) and fit the read data using CountVectorizer
with open('extracted_flu_non_flu_tweets/flu_examples_bag_of_words.txt','r') as f:
    str=f.read()
    vectorizer = CountVectorizer(stop_words='english').fit([str]) #remove stop words in english

#get words and counts in document
word_counts = vectorizer.fit_transform([str]).toarray()[0]
words = vectorizer.get_feature_names()

#pair the words with their word counts as tuples
words_counts=list(zip(words,word_counts))


#sort the tuples in decreasing order of word count
words_counts.sort(key=lambda x : x[1])

words_counts.reverse()

#print the results
print("The top 20 words related to flu-tweets directly realted to personal sickness:")
for i in range(0,21):
    print(words_counts[i][0])



#read the teets (NOT directly related to personal sickness) and fit the read data using CountVectorizer
with open('extracted_flu_non_flu_tweets/non_flu_examples_bag_of_words.txt','r') as f:
    str=f.read()
    vectorizer = CountVectorizer(stop_words='english').fit([str]) #remove stop words in english

#get words and counts in document
word_counts = vectorizer.fit_transform([str]).toarray()[0]
words = vectorizer.get_feature_names()

#pair the words with their word counts as tuples
words_counts=list(zip(words,word_counts))


#sort the tuples in decreasing order of word count
words_counts.sort(key=lambda x : x[1])

words_counts.reverse()

#print the results
print("\nThe top 20 words related to flu-tweets NOT directly realted to personal sickness:")
for i in range(0,21):
    print(words_counts[i][0])
```
